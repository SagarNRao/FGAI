```python
# Step 1: Install the necessary libraries
!pip install -q transformers gradio

# Step 2: Import libraries
from transformers import pipeline
import gradio as gr

# Step 3: Load the sentiment analysis pipeline
# This downloads a pre-trained model automatically (distilbert-base-uncased-finetuned-sst-2-english)
classifier = pipeline("sentiment-analysis")

# Step 4: Define a simple function for the app to use
def analyze_text(text):
    if not text.strip():
        return "Please enter some text!"
    result = classifier(text)[0]
    label = result['label']
    score = round(result['score'], 4)
    return f"Sentiment: **{label}**\nConfidence: **{score}**"

# Step 5: Create the web interface
demo = gr.Interface(
    fn=analyze_text,
    inputs=gr.Textbox(lines=3, placeholder="Write something here...", label="Your sentence"),
    outputs=gr.Textbox(label="Result"),
    title="ðŸ“Š My First AI Sentiment Analyzer",
    description="Type any sentence and see if the AI thinks it's Positive or Negative!",
    examples=[
        "I love this class so much!",
        "This is the worst day ever.",
        "It's okay, nothing special.",
        "Amazing work everyone! ðŸš€"
    ],
    theme="soft",               # nicer look
    allow_flagging="never"
)

# Step 6: Launch the app (share=True gives you a public link)
demo.launch(share=True)
```
